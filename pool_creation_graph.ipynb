### SUSHI

```python
sushi_pools = pd.read_csv("sushi_pairs.csv")
uniq_sushi_pools = set(sushi_pools.id)    #same len as without set
```

```python
sushi_pools_creators = pd.read_csv("sushi_pool_creators.csv")
sushi_pools_creators = sushi_pools_creators.iloc[:, 1:]


sushi_pools_creators.columns = ["id", "creator_address", "creation_txn"]
uniq_sushi_creators = set(sushi_pools_creators.creator_address)
```

```python
sushi_pools = sushi_pools.merge(sushi_pools_creators, on="id")
```

```python
sushi_pools['creator_address'].value_counts()
```

```python
poolsV3['creator_address'].value_counts()
```

```python
common_addrs = uniq_sushi_creators.intersection(uniq_creators)
len(common_addrs)
```

```python
sushi_cc = sushi_pools['creator_address'].value_counts()
uni_cc = poolsV3['creator_address'].value_counts()

common_addrs_counts = []
for add in common_addrs:
    ex = {"id": add,
          "uniswap_pools": uni_cc[add], "sushi_pools": sushi_cc[add]}
    
    common_addrs_counts.append(ex)

common_addrs_counts = pd.DataFrame(common_addrs_counts)
```

```python
common_addrs_counts.sort_values("uniswap_pools", inplace=True)
common_addrs_counts.tail(10)
```
